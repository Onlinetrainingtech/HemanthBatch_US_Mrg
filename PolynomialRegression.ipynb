```python
#Polynomial Regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1.DataCollections
```

```python
dataset=pd.read_csv('Datasets/Position_Salaries.csv')
dataset
```

```python
#separate the input and output
X=dataset.iloc[:,1:2].values
Y=dataset.iloc[:,2].values
```

```python
X
```

```python
Y
```

```python
#3.DataAnalysis data
plt.plot(X,Y)
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,Y)
```

```python
predicated_value=lin_reg.predict(X)
```

```python
plt.scatter(X,Y,color='red')
plt.plot(X,predicated_value,color='blue')
plt.title("Truth or bluff LinearRegression")
plt.xlabel("PositionLevel")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pol_reg=PolynomialFeatures(degree=4)
X_poly=pol_reg.fit_transform(X)
```

```python
lin_reg2=LinearRegression()
lin_reg2.fit(X_poly,Y)
predicated_value=lin_reg2.predict(pol_reg.fit_transform(X))
```

```python
plt.scatter(X,Y,color='red')
plt.plot(X,predicated_value,color='blue')
plt.title('Truth')
plt.xlabel('PositionalLevel')
plt.ylabel('Salary')
plt.show()
```
