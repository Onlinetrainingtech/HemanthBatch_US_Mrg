```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xp=np.array([0,6])
yp=np.array([0,50])
plt.plot(xp,yp)
plt.show()
```

```python
xp=np.array([0,6])
yp=np.array([0,50])
plt.plot(xp,yp,'o')
plt.show()
```

```python
x=np.array([1,2,6,8])
y=np.array([3,8,1,10])
plt.plot(x,y)
plt.show()
```

```python
y=np.array([3,8,1,10])
plt.plot(y,marker='H')
plt.show()
```

```python
y=np.array([3,8,1,10])
x=np.array([10,20,30,40])
plt.plot(x,y,'o:r')
plt.show()
```

```python
x=[10,20,30,40]
y=[50,60,70,80]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4])
plt.plot([10,20,30,40])
plt.plot([1,2,3,4,5],[10,20,30,40,50])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x=[1,2,3,4]
y=[i**2 for i in x]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x=[10,20,30,40]
plt.plot(x,x,'ro')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(0,30)
plt.ylim(0,50)
plt.plot(x)
plt.grid(True)
plt.show()
```

```python
t1=np.arange(5)
plt.plot(t1,'r--',t1**4,'bs',t1**3,'g^',t1**5,'ro')
plt.show()
```

```python
x=[1,2,3,4]
plt.plot(x,[elem for elem in x],label='linear')
plt.plot(x,[elem**2 for elem in x],label='square')
plt.grid(True)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.savefig('myfig.png')
plt.show()
```

```python
#Histogram
data=np.random.randn(8,8)
plt.hist(data)
plt.show()
```

```python
x=[10,20,30,40]
plt.hist(x)
plt.show()
```

```python
x=[10000,20000,30000,40000]
my_label=["a","b","c","d"]
plt.pie(x,labels=my_label)
plt.show()
```

```python
#Scatter plot
x=np.random.rand(100)
y=np.random.rand(100)
plt.scatter(x,y)
plt.show()
```
